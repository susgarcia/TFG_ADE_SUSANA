```python
#Preprocesamiento del texto
nlp = spacy.load("es_core_news_lg")

def preprocess(text):
    # Eliminar tildes y pasar a minúsculas
    text = unidecode(text.lower())

    # Unificar palabras
    text = re.sub(r'\bcabello\b', 'pelo', text)

    doc = nlp(text)
    tokens = []

    for token in doc:
        if (
            not token.is_stop #quitar stopwords
            and not token.is_punct #quitar puntuacion
            and not token.is_digit #quitar numeros
            and token.is_alpha #quitar caracteres especiales
            and token.pos_ != 'PRON' #quitar pronombres
            and token.lemma_.lower() not in stop_words #normalizar en minusculas
        ):
            lemma = token.lemma_.strip().lower()

            # Eliminar "él" si está en el medio
            lemma = lemma.replace(' él', '').replace('él ', '').replace('él', '').strip()

            if lemma:  # Solo añadir si no está vacío
                tokens.append(lemma)

    return tokens
```

```python
import ast
seed(52) #semilla reproducibilidad
#Diccionario a partir de los tokens encontrados
id2word = Dictionary(reviews_limpio['tokens'])
id2word.filter_extremes(no_below=2, no_above=.95)
#Se crea el objeto corpus
corpus = [id2word.doc2bow(d) for d in reviews_limpio['tokens']]
```

```python
#Funcion para encontrar el valor optimo de k
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    coherence_values_topic = []
    model_list_topic = []
    for num_topics in range(start, limit, step):
        model = LdaMulticore(corpus=corpus, num_topics=num_topics, id2word=id2word, passes=5, random_state=42)
        model_list_topic.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values_topic.append(coherencemodel.get_coherence())

    return model_list_topic, coherence_values_topic
```

```python
#Se aplica la función definida en el paso anterior
model_list_topic, coherence_values_topic = compute_coherence_values(dictionary=id2word,
                                                        corpus=corpus,
                                                        texts=reviews_limpio['tokens'],
                                                        start=2, limit=15, step=1) #Establecemos el rango de tópicos
```

```python
#Se crea un csv con los valores coherencia del rango de tópicos establecido en el paso anterior
coherence_values_topic_df = pd.DataFrame(coherence_values_topic)
#Se crea una lista con los valores de las columnas indicados
coherence_values_topic_df=coherence_values_topic_df.iloc[:, 0].tolist()
```

```python
#Ploteamos para buscar K optimo
plt.figure(figsize=(10,7))
estrella = mpath.Path.unit_regular_star(6)
circulo = mpath.Path.unit_circle()
verts = np.concatenate([circulo.vertices, estrella.vertices[::-1, ...] *0.9])
codes = np.concatenate([circulo.codes, estrella.codes])
cut_star = mpath.Path(verts, codes)
plt.plot(range(2, 15), coherence_values_topic_df, '#198A99', marker="o", markersize=10, fillstyle='none', linestyle='--')
plt.axvline(x=3, color='r', linestyle='--')
plt.plot(3, 0.299, 'r', marker=cut_star, markersize=18)
plt.xlabel('Número de tópicos')
plt.ylabel('Índice de coherencia')
plt.show()
```

```python
#Se genera el modelo LDA con el número de k elegido, en este caso k=3 para un mejor entendimiento de los temas
k=3
model_k3 = gensim.models.LdaMulticore(corpus=corpus,
                                        id2word=id2word,
                                        num_topics=k,
                                        passes=20,
                                        iterations=400,
                                        workers=1,
                                        random_state=52)

#Se procede a guardar y cargar todos los resultados obtenidos del modelo
model_k3.save("model_3_topics.model")
```

```python
# Obtener tópico dominante para cada documento
topic_dominant = []
for row in model_k4_load[corpus]:
    row = sorted(row, key=lambda x: x[1], reverse=True)
    topic_dominant.append(row[0][0])

# Crear DataFrame con los tópicos dominantes y las lemas
df_topic = pd.DataFrame({
    'topic_dominant': topic_dominant,
    'lemmas': reviews_limpio['tokens']
})
```

```python
#Definimos la función para crear las gráficas de bigramas y trigramas de cada tópico
def plot_ngrams_by_topic(df_topic, topic_number):
    dft = df_topic[df_topic['topic_dominant'] == topic_number]['lemmas']
    dft = [' '.join(tokens) for tokens in dft if isinstance(tokens, list) and tokens]

    # Vectorizers
    def get_ngrams(docs, n):
        vec = TfidfVectorizer(ngram_range=(n, n), use_idf=True)
        X = vec.fit_transform(docs)
        names = vec.get_feature_names_out()
        scores = X.sum(axis=0).A1
        return dict(zip(names, scores))

    unigrams = get_ngrams(dft, 1)
    bigrams = get_ngrams(dft, 2)
    trigrams = get_ngrams(dft, 3)

    # Plot
    def plot_top_ngrams(ngrams_dict, title, position):
        sorted_items = sorted(ngrams_dict.items(), key=itemgetter(1), reverse=True)[:10]
        plt.subplot(1, 3, position)
        plt.bar([x[0] for x in sorted_items], [x[1] for x in sorted_items])
        plt.xticks(rotation=90)
        plt.title(title)

    plt.figure(figsize=(25, 10))
    plot_top_ngrams(bigrams, f'Tópico {topic_number} – Bigramas', 2)
    plot_top_ngrams(trigrams, f'Tópico {topic_number} – Trigramas', 3)
    plt.tight_layout()
    plt.show()
```
